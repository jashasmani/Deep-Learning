```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/demodata.csv')
print(data.head())

```

```python
plt.scatter(data['Age'], data['Weight'], color='blue')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight')
plt.show()

```

```python

X = data[['Age']].values
y = data['Weight'].values

```

```python
model = LinearRegression()
model.fit(X, y)

```

```python
y_pred = model.predict(X)

```

```python
print(f'Weight (slope): {model.coef_[0]}')
print(f'Bias (intercept): {model.intercept_}')


```

```python
d = 24
predicted_weight = model.predict([[d]])
print(f'Predicted weight for age {d} is {predicted_weight[0]:.2f}')

```

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
